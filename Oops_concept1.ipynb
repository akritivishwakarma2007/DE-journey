```python
class ATM:

  def __init__(self):
    self.pin = ''
    self.balance = 0
    result = self.menu() # Store the returned value
    print(result) # Print the returned value
    print("execute")



  def menu(self):
    userinput = input("""
      hello chose :
      1.set pin
      2.change pin
      3.check balance
      4.withdraw
      5.exit
      """)

    if userinput == '1':
      self.set_pin()
    elif userinput == '2':
      self.change_pin()
    elif userinput == '3':
      self.check_balance()
    elif userinput == '4':
      self.withdraw()
    else:
      self.exit()


  def set_pin(self):
    pin = int(input('enter your pin'))
    self.pin = pin
    balance = int(input("enter you balance"))
    self.balance = balance
    print("pin set successfully")
    self.menu()

  def change_pin(self):
    old_pin = int(input("enter old pin"))
    if old_pin == self.pin:
      new_pin = int(input("enter you new pin"))
      self.pin = new_pin
      print("pin changed successfully")
      self.menu()
    else:
      print("invalid pin")
      self.change_pin()

  def check_balance(self):
    print(f"your balance is : {self.balance}") # Use f-string for printing balance
    self.menu()

  def withdraw(self):
    amount = int(input("enter the amout want to withdraw"))

    if amount < self.balance:
      self.balance = self.balance - amount
      print(f"successfully withdrawl {amount}")
      print(f"balace is now {self.balance}")
      self.menu()
    else:
      print("Insufficient balance or invalid amount.") # Added an else condition for withdraw
      self.menu()

  def exit(self):
    print("exit")
```

```python
obj = ATM()
```

```python
class fraction:

  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,other):
     new_name = self.num*other.den +self.den*other.num
     new_den = self.den*other.den
     return '{}/{}'.format(new_name,new_den)
```

```python
fract = fraction(2,4)
fract2 = fraction(4,2)
```

```python
print(fract+fract2)
```

### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line



```python
class points:

  def __init__(self,x,y):
    self.x_cord= x
    self.y_cord = y

  def __str__(self):
    return f'<{self.x_cord}, {self.y_cord}>'

  def euclidean_ditance(self,other):
    return ((self.x_cord - other.x_cord)**2 + (self.y_cord - other.y_cord)**2)**0.5

  def distance_from_origin(self):
    # return (self.x_cord**2 +self.y_cord**2)**0.5
    return self.euclidean_ditance(points(0,0))
```

```python
P1=points(2,4)
print(P1)
P2 = points(4,2)

print(P1.euclidean_ditance(P2))

P2.distance_from_origin()
```

```python
class line:

  def __init__(self,A,B,C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return f'{self.A}X + {self.B}X + {self.C} = 0'

  def point_on_line(line,point):
    if line.A*point.x_cord + line.B*point.y_cord + line.C == 0:
      return "lies on the line"
    else:
      return "does not lie on the line"
```

```python
l1 = line(1,1,-2)
print(l1)

P1 = points(1,1)
print(P1)

l1.point_on_line(P1)
```
