##Creating Numpy Arrays

```python
#np.array
import numpy as np
a = np.array([1,2,3])
print(a)
print(type(a))
```

```python
#2D 3D array

b = np.array([[1,2,3],[4,5,6]])
print(b)
print(type(b))
```

```python
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
print(type(c))
```

```python
#dtype
np.array([1,2,3], dtype=float)
```

```python
#np.arange
#arange(start,stop,step)
np.arange(1,11)
np.arange(1,21,2)
```

```python
#with reshape
#reshape(row,col)
np.arange(1,11).reshape(5,2)
```

```python
np.arange(16).reshape(2,2,2,2)
```

```python
#np.ones and np.zeros
#np.ones(rows,cols)
np.ones((3,5))
```

```python
#np.zeros(rows,cols)
np.zeros((2,4))
```

```python
#np.random
np.random.random((3,2))
```

```python
#np.linspace
print(np.linspace(-10,10,10))
np.linspace(-10,10,10, dtype= int)
```

```python
#np.identity
np.identity(3)
```

##Array Attributes

```python
a1 = np.arange(11,dtype = np.int32)
a2 = np.arange(12, dtype = float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

print(a1)
print(a2)
print(a3)
```

```python
#ndim -> dimension
a3.ndim
```

```python
#shape
print(a2.shape)
print(a2)
```

```python
#size
print(a2.size)
a2
```

```python
#itemsize
a1.itemsize
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```
