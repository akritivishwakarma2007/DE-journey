##Creating Numpy Arrays

```python
#np.array
import numpy as np
a = np.array([1,2,3])
print(a)
print(type(a))
```

##Array Attributes

```python
#ndim -> dimension
a3.ndim
```

```python
#shape
print(a2.shape)
print(a2)
```

```python
#size
print(a2.size)
a2
```

```python
#itemsize
a1.itemsize
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

##Changing Datatype

```python
#astype
a3.astype(float)
a2.astype(np.int32)
```

##Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
a1
a2
```

```python
#Scalar operations

#Arithematic
print(a1 * 2)
print(a1 + 2)
print(a1 - 2)
print(a1 / 2)
print(a1 ** 2)
```

```python
#relational operators
print(a2>20)
```

```python
#vector Operations
#Arithematic
print(a1 + a2)
print(a1 - a2)
print(a1 * a2)
print(a1 / a2)
```

##Array Fumctions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
#max/min/sum/product
#0 -> cols , 1 -> row
print(np.max(a1,axis = 0))
print(np.min(a1,axis = 1))
print(np.sum(a1,axis = 1))
print(np.prod(a1, axis = 1))
```

```python
#mean/median/std/var
print(np.mean(a1, axis = 0))
print(np.median(a1, axis = 0))
print(np.std(a1, axis = 0))
print(np.var(a1, axis = 0))
```

```python
#trignometric function
np.sin(a1)
```

```python
#dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
#log and exponents
np.exp(a1)
np.log(a1)
```

```python
#round/floor/ceil
print(np.round(np.random.random((2,3))*100))
print(np.floor(np.random.random((2,3))*100))
print(np.ceil(np.random.random((2,3))*100))
```
