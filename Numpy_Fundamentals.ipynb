##Creating Numpy Arrays

```python
#np.array
import numpy as np
a = np.array([1,2,3])
print(a)
print(type(a))
```

```python
#2D 3D array

b = np.array([[1,2,3],[4,5,6]])
print(b)
print(type(b))
```

```python
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
print(type(c))
```

```python
#dtype
np.array([1,2,3], dtype=float)
```

```python
#np.arange
#arange(start,stop,step)
np.arange(1,11)
np.arange(1,21,2)
```

```python
#with reshape
#reshape(row,col)
np.arange(1,11).reshape(5,2)
```

```python
#np.ones and np.zeros
#np.ones(rows,cols)
np.ones((3,5))
```

```python
#np.zeros(rows,cols)
np.zeros((2,4))
```

```python
#np.random
np.random.random((3,2))
```

```python
#np.linspace
np.linspace(-10,10,10)
```

```python
#np.identity
np.identity(3)
```
